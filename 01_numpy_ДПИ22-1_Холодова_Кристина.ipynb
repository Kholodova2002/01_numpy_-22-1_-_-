```python
import numpy as np

data = np.loadtxt('minutes_n_ingredients.csv', delimiter=',', dtype=np.int32)

print("ЗАДАНИЕ 1")
print(data[:5])
print('----------------------')
print("ЗАДАНИЕ 2")
mean = np.mean(data[:, 1:], axis=0)
minimum = np.min(data[:, 1:], axis=0)
maximum = np.max(data[:, 1:], axis=0)
median = np.median(data[:, 1:], axis=0)

# вывод результатов
print("Среднее значение: ", mean)
print("Минимум: ", minimum)
print("Максимум: ", maximum)
print("Медиана: ", median)

print('----------------------')
print("ЗАДАНИЕ 3")
q75 = np.quantile(data[:, 1], 0.75)

# ограничение сверху значений в первом столбце квантилем 0.75
data[:, 1] = np.where(data[:, 1] > q75, q75, data[:, 1])
print(data[:, 1])

print('----------------------')
print("ЗАДАНИЕ 4")

zcount = np.count_nonzero(data[:, 1] == 0)
print('Количество нулей:', zcount)

# подсчет количества рецептов со значением продолжительности, равным нулю
num_zero_duration = data[:, 1].size - np.count_nonzero(data[:, 1])

# замена значений продолжительности, равных нулю, на 1
data[:, 1] = np.where(data[:, 1] == 0, 1, data[:, 1])

print('----------------------')
print("ЗАДАНИЕ 5")
num_unique_recipes = len(np.unique(data[:, 0]))

# вывод количества уникальных рецептов
print("Number of unique recipes:", num_unique_recipes)

print('----------------------')
print("ЗАДАНИЕ 6")

unique_ingredient_counts, counts = np.unique(data[:, 2], return_counts=True)
num_unique_ingredient_counts = len(unique_ingredient_counts)

# вывод количества и различных значений количества ингредиентов
print("Number of unique ingredient counts:", num_unique_ingredient_counts)
print("Unique ingredient counts:", unique_ingredient_counts)
print("Counts of each unique ingredient count:", counts)

print('----------------------')
print("ЗАДАНИЕ 7")

# создание логического массива для выбора рецептов, содержащих не более 5 ингредиентов
less_than_5_ingredients = data[:, 2] <= 5

# выбор соответствующих строк из исходного массива
filtered_data = data[less_than_5_ingredients]

# вывод первых 5 строк нового массива
print(filtered_data[:5])

print('----------------------')
print("ЗАДАНИЕ 8")
# вычисление количества ингредиентов, которые приходятся на одну минуту рецепта
ingredients_per_minute = data[:, 2] / data[:, 1]

# нахождение максимального значения
max_value = np.max(ingredients_per_minute)

# вывод максимального значения
print("Максимальное количество ингредиентов, которые приходятся на одну минуту рецепта: ", max_value)

print('----------------------')
print("ЗАДАНИЕ 9")
sorted_data = data[data[:,1].argsort()[::-1]]

# выбор первых 100 рецептов
top_100_data = sorted_data[:100]

# вычисление среднего значения количества ингредиентов для топ-100 рецептов
mean_ingredients = np.mean(top_100_data[:,2])

# вывод среднего значения на экран
print("Среднее количество ингредиентов для топ-100 рецептов с наибольшей продолжительностью:", mean_ingredients)
print('----------------------')
print("ЗАДАНИЕ 10")
# генерация случайных индексов
random_indices = np.random.choice(data.shape[0], size=10, replace=False)

# выбор данных с помощью случайных индексов
random_data = data[random_indices]

# вывод информации о рецептах на экран
for recipe in random_data:
    print("Идентификатор рецепта:", recipe[0])
    print("Продолжительность выполнения в минутах:", recipe[1])
    print("Количество ингредиентов:", recipe[2])
print('----------------------')
print("ЗАДАНИЕ 11")
# вычисление среднего количества ингредиентов в рецептах
mean_ingredients = np.mean(data[:, 2])

# подсчет количества рецептов с количеством ингредиентов меньше среднего
less_than_mean = np.sum(data[:, 2] < mean_ingredients)

# вычисление процента рецептов с количеством ингредиентов меньше среднего
percent_less_than_mean = less_than_mean / data.shape[0] * 100

# вывод результата на экран
print("Процент рецептов, количество ингредиентов в которых меньше среднего: {:.2f}%".format(percent_less_than_mean))
print('----------------------')
print("ЗАДАНИЕ 12")
# создание нового массива с информацией о длительности выполнения и количестве ингредиентов
time_and_ingr = data[:, 1:]

# создание нового столбца, содержащего информацию о простоте рецепта
simple_recipe = np.where((time_and_ingr[:, 0] <= 20) & (time_and_ingr[:, 1] <= 5), 1, 0)

# объединение исходного массива и нового столбца
data_with_simple_recipe = np.hstack((data, simple_recipe.reshape(-1, 1)))

# вывод первых 5 строк нового массива
print(data_with_simple_recipe[20:30, :])
print('----------------------')
print("ЗАДАНИЕ 13")
simple_count = np.sum(data_with_simple_recipe[:, 3]) # количество "простых" рецептов
total_count = data_with_simple_recipe.shape[0] # общее количество рецептов
simple_percent =  simple_count / total_count * 100 # процент "простых" рецептов
print("Процент 'простых' рецептов: {:.2f}%".format(simple_percent))

print('----------------------')
print("ЗАДАНИЕ 14")
# создаем маски для каждой группы рецептов
short_mask = data[:, 1] < 10
standard_mask = (data[:, 1] > 10) & (data_with_simple_recipe[:, 1] < 20)
long_mask = data[:, 1] >= 20

# выбираем максимальное количество рецептов из каждой группы
max_short = np.where(short_mask)[0].shape[0]
max_standard = np.where(standard_mask)[0].shape[0]
max_long = np.where(long_mask)[0].shape[0]

# создаем трехмерный массив
grouped_recipes = np.zeros((3, max_long, 3), dtype=np.int32)

# заполняем трехмерный массив
grouped_recipes[0, :max_short, :] = data[short_mask][:max_short]
grouped_recipes[1, :max_standard, :] = data[standard_mask][:max_standard]
grouped_recipes[2, :max_long, :] = data[long_mask][:max_long]

# выводим форму массива
print(grouped_recipes.shape)




```
